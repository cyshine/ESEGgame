## Importação do ngrok no colab
Importa a biblioteca ngrok no colab e configura para direcionar a porta local 8051 (porta padrão do streamlit) para a porta do ngrok.

Esta linha exibe uma url, que tornará a aplicação do streamlit acessível externamente ao ambiente virtual do colab (de qualquer máquina, pela internet).

```python
from pyngrok import ngrok
```

```python
public_url = ngrok.connect('5000')
public_url = str(public_url).split("\"")[1]
with open("public_url.txt", "w") as f: 
    f.write(str(public_url))

print(public_url)
```
